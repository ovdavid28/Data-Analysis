Function that create a DataFrame giving it the year

```python
def by_year(year):
    df=pd.DataFrame({
        'Population': population[str(year)],
        'Fertility': fertility[str(year)],
        'Life expectancy': life_exp[str(year)],
        'Group': regions['Group'],
    })
    df["ID"] = np.arange(len(df))
    return df
```

# 2.Create a function update_year that receives a year as an integer and updates the data of the existing figure with the values from the selected year. Note: The update might not be very efficient

```python
def update_year(years):
    fig = go.FigureWidget()
    df=by_year(years)
    for group_name, sub_df in df.groupby("Group"):
        sc = [fig.add_scatter(
            x=sub_df['Fertility'],
            y=sub_df['Life expectancy'],
            mode='markers',
            marker={
                'size': np.sqrt(sub_df['Population'].fillna(0))/ 400
            },
            name=group_name,opacity=0.6,
        )]
    return fig
```

# 4.Bind the update_year function to changes in the horizontal slider and use it to interactively change the plot

I have tried many different ways in order to update the graph and doing it smoothly, but after many tries and not finding the right solution I did a wide search on google and found out this code from a question that you made. I went through the code and understood really well how it works.
https://github.com/plotly/plotly.py/issues/1288

```python
fig = go.FigureWidget()
fig.layout.yaxis.range = [20, 90]
fig.layout.xaxis.range = [0, 10]
#fig.layout.title = "Fertility vs Life_expectancy:{}".format(update_year(years))
fig['layout']['xaxis'] = {'title': 'Fertility', 'gridcolor': '#FFFFFF'}
fig['layout']['yaxis'] = {'title': 'Life expectancy', 'type': 'log', 'gridcolor': '#FFFFFF'}


scs = {}
for group_name, subdf in df.groupby("Group"):
    sc = fig.add_scatter(
        x=subdf["Fertility"],
        y=subdf["Life expectancy"],
        ids=subdf['ID'],
        mode='markers',
        marker={
            'size': np.sqrt(subdf["Population"].fillna(1) / np.pi) / 200
        },
        text=subdf.index,
        name=group_name,
    )
    scs[group_name] = sc
slider = IntSlider(
    description="Years",
    value=min(years.astype(int)), 
    min=min(years.astype(int)), 
    max=max(years.astype(int)),
    continuous_update=False)

def update_year(years):
    df=by_year(years)
    fig.layout.title = "Fertility vs Life_expectancy {}".format(years)
    with fig.batch_animate(duration=750, easing='cubic-in-out'):
        for group_name, subdf in df.groupby("Group"):
            scs[group_name].x = subdf["Fertility"]
            scs[group_name].y = subdf["Life expectancy"]
            ids=subdf["ID"],
            scs[group_name].marker.size = np.sqrt(subdf["Population"].fillna(1) / np.pi) / 200
            

interact(update_year, years=slider)
fig
```

What we can see clearly in the evolution of this graph is that the fertility index decreases and life expectancy increases over the years for mainly all the continents, being Sub-Saharan Africa the one with less increase over the years.
